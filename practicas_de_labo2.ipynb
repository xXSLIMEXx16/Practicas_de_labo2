Crea un programa que encuentre todos los números perfectos en un rango dado.
Un número perfecto es un número que es igual a la suma de sus divisores propios
positivos (excluyendo el número mismo)

```python
numero = int(input('Ingrese un numero : '))
numeros_perfectos = []

for i in range(1 , numero + 1 ):
  divisores = 0
  suma_divisores = 0

  for divisores in range(1, i):
    if i % divisores == 0:
      suma_divisores += divisores

  if suma_divisores == i:
    numeros_perfectos.append(i)

print(numeros_perfectos)
```

Escribe una función llamada `crear_diccionario` que tome una lista de claves
y una lista de valores, y retorne un diccionario donde cada clave esté asociada
a su valor correspondiente.

```python
def crear_diccionario(datos):
  diccionario = {}
  for clave, valor in datos:
    diccionario[clave] = valor
  return diccionario

lista_datos = [('Nombre', 'Mario'),
               ('Apellido', 'Arana'),
               ('Edad', '24 años'),
               ('Pais', 'Argentina')]
diccionario = crear_diccionario(lista_datos)

print(diccionario)
```

nuevas claves y valores, agregando las nuevas claves si no existen

```python
def actualizar_diccionario(clave, valor, mi_diccionario):
  for clave, valor in zip(clave, valor):
    mi_diccionario[clave]=valor
  return mi_diccionario
diccionario = {'Nombre': 'Mario', 'Apellido': 'Arana', 'Edad': '24 años', 'Pais': 'Argentina'}
print(diccionario)
lista_claves = ['Nombre', 'Apellido','Edad','Pais']
lista_datos = ['Eduardo', 'Gomez','20 años', 'Bolivia']

diccionario = actualizar_diccionario(lista_claves, lista_datos, diccionario)
print(diccionario)
```

entre los dos puntos

```python
from IPython.core.magics.execution import math
def distancia(coordenada1, coordenada2):
  return math.sqrt((coordenada1[0]-coordenada2[0])**2 + (coordenada1[1]-coordenada2[1])**2)
coordenada1 = (25, 30)
coordenada2 = (10, 5)

print('La distancia entre las coordenadas es: ', round(distancia(coordenada1, coordenada2),2) )
```

Escribe una función llamada `estadisticas_notas` que tome un diccionario
donde las claves son nombres de estudiantes y los valores son listas de notas.
La función debe retornar un diccionario donde las claves son los nombres y los
valores son tuplas con el promedio y la nota máxima de cada estudiante.

```python
def estadisticas_notas(**kwargs):
    resultado = {}
    for estudiante, notas in kwargs.items():
        promedio = sum(notas) / len(notas)
        maxima = max(notas)
        resultado[estudiante] = (promedio, maxima)
    return resultado

notas_estudiantes = {
 'Ana': [7, 4, 9, 10, 6],
 'Luis': [8, 5, 7, 9, 6],
 'Carla': [10, 9, 6, 8, 7],
 'Diego': [5, 6, 9, 7, 8],

}

print(estadisticas_notas(**notas_estudiantes))
```

un valor, y retorne un nuevo diccionario que solo contenga los pares clave-valor
donde el valor sea mayor que el valor dado

```python
def filtrar_diccionario(diccionario, resultado):
  nuevo_dic = {}
  for clave, valor in diccionario.items():
    if valor>resultado and diccionario[clave] % 2 == 0:
      nuevo_dic[clave] = valor
  return nuevo_dic



mi_diccionario = {
 'Ana': 4,
 'Luis': 5,
 'Carla': 6,
 'Diego': 9,
}

filtrar_diccionario(mi_diccionario,3)
```

con las mismas claves y retorne un nuevo diccionario donde los valores sean la

```python
def sumar_diccionarios(diccionario1, diccionario2):
  diccionario_nuevo = {}
  for clave, valor in diccionario1.items():
    diccionario_nuevo[clave] = (diccionario1[clave] + diccionario2[clave])
  return diccionario_nuevo

diccionario1 = {
    "Enero": 100,
    "Febrero": 150,
    "Marzo": 200
}

diccionario2 = {
    "Enero": 120,
    "Febrero": 180,
    "Marzo": 210
}

nuevo_diccionario = sumar_diccionarios(diccionario1, diccionario2)
print(nuevo_diccionario)
```
