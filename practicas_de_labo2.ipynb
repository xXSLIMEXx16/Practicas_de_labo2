positivos (excluyendo el número mismo)

```python
numero = int(input('Ingrese un numero : '))
numeros_perfectos = []

for i in range(1 , numero + 1 ):
  divisores = 0
  suma_divisores = 0

  for divisores in range(1, i):
    if i % divisores == 0:
      suma_divisores += divisores

  if suma_divisores == i:
    numeros_perfectos.append(i)

print(numeros_perfectos)
```

Escribe una función llamada `crear_diccionario` que tome una lista de claves
y una lista de valores, y retorne un diccionario donde cada clave esté asociada
a su valor correspondiente.

```python
def crear_diccionario(datos):
  diccionario = {}
  for clave, valor in datos:
    diccionario[clave] = valor
  return diccionario

lista_datos = [('Nombre', 'Mario'),
               ('Apellido', 'Arana'),
               ('Edad', '24 años'),
               ('Pais', 'Argentina')]
diccionario = crear_diccionario(lista_datos)

print(diccionario)
```

Escribe una función llamada `estadisticas_notas` que tome un diccionario
donde las claves son nombres de estudiantes y los valores son listas de notas.
La función debe retornar un diccionario donde las claves son los nombres y los
valores son tuplas con el promedio y la nota máxima de cada estudiante.

```python
def estadisticas_notas(**kwargs):
    resultado = {}
    for estudiante, notas in kwargs.items():
        promedio = sum(notas) / len(notas)
        maxima = max(notas)
        resultado[estudiante] = (promedio, maxima)
    return resultado

notas_estudiantes = {
 'Ana': [7, 4, 9, 10, 6],
 'Luis': [8, 5, 7, 9, 6],
 'Carla': [10, 9, 6, 8, 7],
 'Diego': [5, 6, 9, 7, 8],

}

print(estadisticas_notas(**notas_estudiantes))
```

donde el valor sea mayor que el valor dado

```python
def filtrar_diccionario(diccionario, resultado):
  nuevo_dic = {}
  for clave, valor in diccionario.items():
    if valor>resultado and diccionario[clave] % 2 == 0:
      nuevo_dic[clave] = valor
  return nuevo_dic



mi_diccionario = {
 'Ana': 4,
 'Luis': 5,
 'Carla': 6,
 'Diego': 9,
}

filtrar_diccionario(mi_diccionario,3)
```

Desarrolla un sistema de autenticación que gestione el inicio de sesión de usuarios. Usa un decorador para verificar las credenciales del usuario antes de permitirle acceder a una función protegida (por ejemplo, una función que muestre
un mensaje de bienvenida). Implementa un decorador adicional para limitar el
número de intentos fallidos de inicio de sesión.

```python
# Base de datos simple
USUARIOS = {
    "mario": "1234",
    "juan": "abcd"
}

# Decorador para limitar intentos
def limitar_intentos(max_intentos):
    def decorador(func):
        intentos = 0

        def wrapper(*args, **kwargs):
            nonlocal intentos
            if intentos >= max_intentos:
                print(" Se superó el número máximo de intentos.")
                return
            resultado = func(*args, **kwargs)
            if resultado is False:
                intentos += 1
            return resultado

        return wrapper
    return decorador

# Decorador para verificar credenciales
def verificar_credenciales(func):
    def wrapper(*args, **kwargs):
        usuario = input("Usuario: ")
        password = input("Contraseña: ")

        if usuario in USUARIOS and USUARIOS[usuario] == password:
            return func(usuario, *args, **kwargs)
        else:
            print("Credenciales incorrectas.\n")
            return False   # para que el decorador de intentos sume
    return wrapper


@limitar_intentos(3)
@verificar_credenciales
def bienvenida(usuario):
    print(f"Bienvenido {usuario}, acceso concedido.")

bienvenida()
```

Crea un programa con varias funciones que realicen operaciones matemáticas
(suma, resta, multiplicación y división). Implementa un decorador @logger que
registre cada llamada a estas funciones, imprimiendo el nombre de la función,
los argumentos y el resultado cada vez que se ejecute alguna de las funciones.


```python
def logger(func):
    def wrapper(*args, **kwargs):
        print(f"Llamando a: {func.__name__}")
        print(f" Argumentos: {args}")

        resultado = func(*args, **kwargs)

        print(f" Resultado: {resultado}")
        return resultado
    return wrapper


@logger
def suma(a, b):
    return a + b

@logger
def resta(a, b):
    return a - b

@logger
def multiplicacion(a, b):
    return a * b

@logger
def division(a, b):
    if b == 0:
        return "Error: división por cero"
    return a / b


# Ejemplos de uso
suma(4, 6)
resta(10, 3)
multiplicacion(7, 5)
division(20, 4)
division(10, 0)
```

```python
class Matriz:
  def __init__(self, ma_triz):
    self.datos = np.array(ma_triz)

  def transpuesta(self):
    print(f"La matriz trasnpuesta es: \n{np.transpose(self.datos)}")

  def determinante(self):
    print(f"El determinante de la matriz es: \n {np.linalg.det(self.datos)}")

  def inversa(self):
    print(f"La matriz inversa es: \n {np.linalg.inv(self.datos)}")
matriz_3d = ([[2,5,8],
              [9,2,3],
              [4,7,6]])

mi_matriz = Matriz(matriz_3d)
mi_matriz.transpuesta()
mi_matriz.determinante()
mi_matriz.inversa()
```
